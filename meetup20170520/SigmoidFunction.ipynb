```python
#import numpy for using decimal places
import numpy as np
#import pyplot for decribing
import matplotlib.pyplot as plt

#sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
#from -10.0 to 10.0 , 1.0 Shredded 
x = np.arange(-10.0, 10.0, 1.0)
y = sigmoid(x)

#label
plt.xlabel('input')
plt.ylabel('output')
plt.plot(x, y)
#y-axis-range
plt.ylim(-0.1, 1.1)
plt.show()
```
